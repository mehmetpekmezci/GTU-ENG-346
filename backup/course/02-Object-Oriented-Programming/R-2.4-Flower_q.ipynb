Write a Python class, Flower, that has three instance variables of type str, int, and float, that respectively represent the name of the flower, its number of petals, and its price. Your class must include a constructor method that initializes each variable to an appropriate value, and your class should include methods for setting the value of each type, and retrieving the value of each type.



```python
class Flower:

    def __init__(self, name, petals, price):
        self.__name = name
        self.__petals = petals
        self.__price = price

    def display_info(self):
        print ("Flower: {}, Petals: {}, Price: {}".format(self.__name, self.__petals, self.__price))


    def get_name(self):
        return self.__name

    def get_petals(self):
        return self.__petals

    def get_price(self):
        return self.__price

    def set_name(self, name):
        self.__name = name

    def set_petals(self, petals):
        self.__petals = petals

    def set_price(self, price):
        self.__price = price
```

```python
sumbul = Flower("Sumbul", 7, 12.50)
```

```python
sumbul.display_info()
```

```python
sumbul.get_name()
```

```python
sumbul.set_name('Sumbull')
```

```python
sumbul.display_info()
```

```python
sumbul.set_price(13.0)
```

```python
class Flower2:

    def __init__(self):
        self.attr_list = {}

    def display_info(self):
        for k in self.attr_list:
            print (k, self.attr_list[k])

    def get_attr(self, attr):
        return self.attr_list[attr]

    def set_attr(self, attr, value):
        self.attr_list[attr] = value

```

```python
gul = Flower2()
gul.set_attr('name', 'Gul')
gul.set_attr('petals', 8)
gul.set_attr('price', 95)

gul.display_info()
```
