# Inheritance

```python
class Vehicle(object):
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        pass

    def stop_engine(self):
        pass

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model
```

```python
class Car(Vehicle):
    def __init__(self, brand, model, color):
        super().__init__(brand, model)
        self.__color = color

    def start_engine(self):
        return "Car engine started."

    def stop_engine(self):
        return "Car engine stopped."

    def get_color(self):
        return self.__color

    def set_color(self, color):
        self.__color = color
```

```python
class Bus(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)
        self.__seats = seats

    def start_engine(self):
        return "Bus engine started."

    def stop_engine(self):
        return "Bus engine stopped."

    def get_seats(self):
        return self.__seats

    def set_seats(self, seats):
        self.__seats = seats
```

```python
class Motorcycle(Vehicle):
    def __init__(self, brand, model, wheels):
        super().__init__(brand, model)
        self.__wheels = wheels

    def start_engine(self):
        return "Motorcycle engine started."

    def stop_engine(self):
        return "Motorcycle engine stopped."

    def open_foot_brace(self):
        return "Motorcycle is on foot."

    def close_foot_brace(self):
        return "Motorcycle is off foot."

    def get_wheels(self):
        return self.__wheels

    def set_wheels(self, wheels):
        self.__wheels = wheels

    def display_info(self):
        return f"Brand: {self.get_brand()}, Model: {self.get_model()}, Wheels: {self.__wheels}\nCycling is adventurous."
```

```python
class Bicycle(Vehicle):
    def __init__(self, brand, model, wheels):
        super().__init__(brand, model)
        self.__wheels = wheels

    def start_engine(self):
        return "Bicycles do not have engines."

    def stop_engine(self):
        return "Bicycles do not have engines."

    def open_foot_brace(self):
        return "Bicycle is on foot."

    def close_foot_brace(self):
        return "Bicycle is off foot."

    def get_wheels(self):
        return self.__wheels

    def set_wheels(self, wheels):
        self.__wheels = wheels

    def display_info(self):
#       return f"Brand: {self.__brand}, Model: {self.__model}, Wheels: {self.__wheels}\nCycling is healthy."
        return f"Brand: {self.get_brand()}, Model: {self.get_model()}, Wheels: {self.__wheels}\nCycling is healthy."

```

```python
car = Car("Toyota", "Camry", "green")
bus = Bus("Volvo", "9600", 48)
motorcycle = Motorcycle("Harley-Davidson", "Sportster", 3)
bicycle = Bicycle("Schwinn", "Mountain Bike", 3)
```

```python
print(car.display_info())
print(car.start_engine())
print(car.stop_engine())
```
