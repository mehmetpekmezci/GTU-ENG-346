# Encapsulation

```python
class Vehicle:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        pass

    def stop_engine(self):
        pass

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, brand):
        self.__brand = model
```

```python
class Car:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        return "Car engine started."

    def stop_engine(self):
        return "Car engine stopped."

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model
```

```python
class Bus:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        return "Bus engine started."

    def stop_engine(self):
        return "Bus engine stopped."

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, brand):
        self.__brand = model
```

```python
class Motorcycle:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        return "Motorcycle engine started."

    def stop_engine(self):
        return "Motorcycle engine stopped."

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, brand):
        self.__brand = model
```

```python
class Bicycle(Vehicle):
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model

    def start_engine(self):
        return "Bicycles do not have engines."

    def stop_engine(self):
        return "Bicycles do not have engines."

    def display_info(self):
        return f"Brand: {self.__brand}, Model: {self.__model}"

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, brand):
        self.__brand = model
```

```python
car = Car("Toyota", "Camry")
bus = Bus("Volvo", "9600")
motorcycle = Motorcycle("Harley-Davidson", "Sportster")
bicycle = Bicycle("Schwinn", "Mountain Bike")
```

```python
print(car.display_info())
print(car.start_engine())
print(car.stop_engine())
```

```python
print(bus.display_info())
print(bus.start_engine())
print(bus.stop_engine())
```

```python
print(motorcycle.display_info())
print(motorcycle.start_engine())
print(motorcycle.stop_engine())
```

```python
print(bicycle.display_info())
print(bicycle.start_engine())
print(bicycle.stop_engine())
```
