R-1.1 Write a short Python function, *is_multiple(n, m)*, that takes two integer values and returns **True** if *n* is a multiple of *m*, that is, *n = mi* for some integer *i*, and **False** otherwise.

```python
# Answer for R-1.1 on page 51
def is_multiple(n, m):
    """
    This function returns True if n = mi for some integer i. Otherwise it returns False.
    n: multiple of m to check
    """
    ret = False

    ret = (n % m == 0)

    return ret
```

determines if there is a distinct pair of numbers in the sequence whose
product is odd.

```python
def odd_pairs(list_of_numbers):
    l = len(list_of_numbers)
    pair_list = []

    odd_numbers = []
    for i in range(l):
        if list_of_numbers[i] % 2 == 1:
            odd_numbers.append(list_of_numbers[i])

    for i in range(len(odd_numbers)):
        for j in range(i+1, len(odd_numbers)):
            pair_list.append((odd_numbers[i], odd_numbers[j]))


    return pair_list
```
