```python
class Node:
    def __init__(self, data):
        self.data= data
        self.left = None
        self.right = None
# self.parent = None

    def __str__(self):
        # return "{}|{}|{}".format(self.data, self.left, self.right)
        return "{}".format(self.data)
```

```python
class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):

        if self.root == None:
            node = Node(data)
            self.root = node
        else:
            current = self.root
            while True:
                # print("current:", current)
                prev = current
                if data < current.data:
                    current = current.left
                else:
                    current = current.right

                if current == None:
                    node = Node(data)
# node.parent = prev
                    if data < prev.data:
                        prev.left = node
                    else:
                        prev.right = node

                    break

    def display_visit(self, node, depth):
        if node == None: return

        print ('    '*depth, end='')
        print (node.data)

        self.display_visit(node.left, depth+1)
        self.display_visit(node.right, depth+1)


    def display(self):
        self.display_visit(self.root, 0)

    def in_order_visit(self, node):

        if node == None: return

        self.in_order_visit(node.left)
        print (node, "->", end='')
        self.in_order_visit(node.right)


    def in_order(self):
        self.in_order_visit(self.root)
        print()

    def search_visit(self, node, data):
        if node == None: return None

        if node.data == data: return node

        if data < node.data:
            return self.search_visit(node.left, data)
        else:
            return self.search_visit(node.right, data)

    def recursive_search(self, data):
        return self.search_visit(self.root, data)

    def iterative_search(self, data):

        current = self.root

        while True:
            if current == None: return None
            if current.data == data: return current

            if data < current.data:
                current = current.left
            else:
                current = current.right

```
