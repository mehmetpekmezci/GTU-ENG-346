```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def __str__(self):
        return self.data
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def insert_to_front(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.count += 1

    def append_to_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            current = self.tail
            current.next = new_node
            self.tail = new_node
        self.count += 1

    def remove_from_front(self):
        current = self.head
        self.head = self.head.next
        current.next = None
        self.count -= 1
        return current

    def display(self):
        print("Len:", self.count)
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

```

```python
ll = SinglyLinkedList()
```

```python
ll.insert_to_front(5)
ll.insert_to_front(4)
ll.display()
```

```python
ll.remove_from_front()
ll.remove_from_front()
ll.display()
```

```python
class Stack:

    def __init__(self):
        self.ll = SinglyLinkedList()

    def push(self, data):
        self.ll.insert_to_front(data)

    def pop(self):
        if self.is_empty():
            pass
        node = self.ll.remove_from_front()
        return node.data

    def is_empty(self):
        return self.ll.count == 0

    def top(self):
        if self.is_empty():
            pass
        return self.ll.head.data

    def __len__(self):
        return self.ll.count

    def display(self):
        self.ll.display()
```

```python
S = Stack()
```

```python
S.push(5)
```

```python
S.display()
```

```python
S.push(10)
S.push(4)
S.push(101)
S.display()
```

```python
S.pop()
```

```python
S.top()
```

```python
S.display()
```

```python
class Queue:

    def __init__(self):
        self.ll = SinglyLinkedList()

    def enqueue(self, data):
        self.ll.append_to_end(data)

    def dequeue(self):
        node = self.ll.remove_from_front()
        return node.data

    def __len__(self):
        return self.ll.count

    def first(self):
        return self.ll.head.data

    def is_empty(self):
        return self.ll.count == 0

    def display(self):
        self.ll.display()
```

```python
Q = Queue()
```

```python
Q.enqueue(5)
Q.display()
```

```python
Q.enqueue(7)
Q.display()
```

```python
Q.enqueue(9)
Q.enqueue(1)
Q.enqueue(-2)
Q.display()
```

```python
Q.dequeue()
```

```python
Q.display()
```
