```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def __str__(self):
        return self.data
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append_to_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            current = self.tail
            current.next = new_node
            self.tail = new_node
        self.count += 1

    def insert_to_front(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.count += 1

    def remove_from_front(self):
        current = self.head
        self.head = self.head.next
        current.next = None
        self.count -= 1
        return current

    def insert_after_node(self, node, data):
        new_node = Node(data)
        new_node.next = node.next
        node.next = new_node
        if node == self.tail: self.tail = new_node
        self.count += 1

    def insert_to_index(self, i, data):
        assert (i < self.count)

        current = self.head
        j = 0

        while j < i:
            current = current.next
            j += 1

        self.insert_after_node(current, data)

    def find_min(self):
        if self.head == None:
            pass
        min_value = self.head.data

        current = self.head

        while current:
            min_value = min(min_value, current.data)
            current = current.next

        return min_value

    def display(self):
        print("Len:", self.count)
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

```

```python
# Creating a singly linked list
sll = SinglyLinkedList()
```

```python
sll.append_to_end(1)
sll.display()
sll.append_to_end(2)
sll.display()
sll.append_to_end(3)

# Displaying the linked list
sll.display()
```
